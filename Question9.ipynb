```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.5, 100, 0.5)

y1 = x
y2 = x**2
y3 = x**3 / 100
y4 = np.sin(x)
y5 = np.sin(x) / x
y6 = np.log(x)
y7 = np.exp(x)

plt.figure(figsize=(10, 10))

plt.subplot(3, 3, 1)
plt.plot(x, y1, label='y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x')
plt.legend()

plt.subplot(3, 3, 2)
plt.plot(x, y2, label='y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^2')
plt.legend()

plt.subplot(3, 3, 3)
plt.plot(x, y3, label='y = x^3 / 100')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^3 / 100')
plt.legend()

plt.subplot(3, 3, 4)
plt.plot(x, y4, label='y = sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sin(x)')
plt.legend()

plt.subplot(3, 3, 5)
plt.plot(x, y5, label='y = sin(x) / x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sin(x) / x')
plt.legend()

plt.subplot(3, 3, 6)
plt.plot(x, y6, label='y = log(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = log(x)')
plt.legend()

plt.subplot(3, 3, 7)
plt.plot(x, y7, label='y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = e^x')
plt.legend()

plt.tight_layout()

plt.show()
```
